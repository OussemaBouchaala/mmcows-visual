```python
# Cell 2: Create the folder structure
import os

# Define the root folder for your dataset
dataset_root = "custom_dataset"

# Define the subdirectories
subdirs = [
    'train/images', 'train/labels',
    'val/images', 'val/labels',
    'test/images', 'test/labels'
]

for subdir in subdirs:
    os.makedirs(os.path.join(dataset_root, subdir), exist_ok=True)

print(f"Successfully created directory tree at: {os.path.abspath(dataset_root)}")
```

```python
# Cell 3: Check YAML configuration
import yaml
from pathlib import Path 

DATASET_ROOT = Path(r"C:\Users\DELL\Desktop\test\mmcows\data\raw\visual_data")    

yaml_path = DATASET_ROOT / 'yolo_nano/mmcows_binary.yaml'

with open(yaml_path, 'r') as f:
    config = yaml.safe_load(f)

print("--- YAML Content ---")
print(yaml.dump(config, default_flow_style=False))

# Quick check on classes
num_classes = config.get('nc', 0)
class_names = config.get('names', [])
print(f"\nTraining for {num_classes} classes: {class_names}")
```

```python
import torch
# Get the name of the GPU and total memory
gpu_name = torch.cuda.get_device_name(0)
total_memory = torch.cuda.get_device_properties(0).total_memory / 1e9 # Convert to GB
print(f"GPU: {gpu_name}")
print(f"Total VRAM: {total_memory:.2f} GB")
```

## Training

```python
# Cell 4: Start Training
from ultralytics import YOLO

# 1. Load the pre-trained YOLOv8 Nano model
model = YOLO('yolov8n.pt') 

# 2. Run the training process
# We use the path to your yaml file
results = model.train(
    data=yaml_path,
    epochs=100,
    imgsz=640,
    batch=16,          # You can increase this to 32 if your GPU has high VRAM
    device=0,          # Uses your GPU
    amp=False,
    name='cow_detection_v8n',
    exist_ok=True      # Overwrites the folder if you run this twice
)
```
